## Loop Through All 242 Pokemon

```python
# arceus.gg website:

url = 'https://www.arceus.gg/'
```

```python
# Open up arceus.gg website:

driver = webdriver.Chrome('C:\Program Files (x86)\chromedriver.exe')
driver.get(url)

# Maximize Window:
driver.maximize_window()
driver.implicitly_wait(5)
```

```python
## Scrape information from arceus.gg

# Initialize lists:
number_list = []
pokemon_list = []
poke_type_one = []
poke_type_two = []
hp_list = []
attack_list = []
defense_list = []
sp_atk_list = []
sp_def_list = []
speed_list = []
total_stats_list = []

# Pokemon links
pokemon_links = [x.get_attribute("href") for x 
                 in driver.find_elements_by_xpath('//*[@class = "pokemon__list__link"]')]

for link in pokemon_links:
    driver.implicitly_wait(3)

    try:
        driver.get(link)
        driver.implicitly_wait(2)
    except:
        None
        
    # Get Pokemon Numbers
    try:
        num = driver.find_element_by_xpath('//*[@class="pokemon__number"]')
        number_list.append(num.text.replace('#', ''))
    except:
        number_list.append('N/A')
    driver.implicitly_wait(2)
    
    # Get Pokemon Name:
    try:
        name = driver.find_element_by_xpath("//*[@class='app__page__title']")
        pokemon_list.append(name.text.replace('Pokémon Legends: Arceus ', ''))
    except:
        pokemon_list.append('N/A')
    driver.implicitly_wait(2)
    
    # Pokemon type 1:
    try:
        type_one = driver.find_element_by_xpath('//*[@class="pokemon__info__types"]/div[1]/h3').text
        poke_type_one.append(type_one)
    except:
        poke_type_one.append('N/A')
    driver.implicitly_wait(2)
    
    # Pokemon type 2:
    try:
        type_two = driver.find_element_by_xpath('//*[@class="pokemon__info__types"]/div[2]/h3').text
        poke_type_two.append(type_two)
    except:
        poke_type_two.append('N/A')
    driver.implicitly_wait(2)
    
    # HP
    try:
        hp = driver.find_element_by_xpath('//*[@class="pokemon__stat__key"][contains(text(), "HP")]/following-sibling::div[2]')
        hp_list.append(hp.text)
    except:
        hp_list.append('N/A')
    driver.implicitly_wait(2)
    
    # Attack
    try:
        atk = driver.find_element_by_xpath('//*[@class="pokemon__stat__key"][contains(text(), "Attack")]/following-sibling::div[2]')
        attack_list.append(atk.text)
    except:
        attack_list.append('N/A')
    driver.implicitly_wait(2)
    
    # Defense
    try:
        defense = driver.find_element_by_xpath('//*[@class="pokemon__stat__key"][contains(text(), "Defense")]/following-sibling::div[2]')
        defense_list.append(defense.text)
    except:
        defense_list.append('N/A')
    driver.implicitly_wait(2)
    
    # Special Attack
    try:
        sp_atk = driver.find_element_by_xpath('//*[@class="pokemon__stat__key"][contains(text(), "Sp. Atk")]/following-sibling::div[2]')
        sp_atk_list.append(sp_atk.text)
    except:
        sp_atk_list.append('N/A')
    driver.implicitly_wait(2)
    
    # Special Defense
    try:
        sp_def = driver.find_element_by_xpath('//*[@class="pokemon__stat__key"][contains(text(), "Sp. Def")]/following-sibling::div[2]')
        sp_def_list.append(sp_def.text)
    except:
        sp_def_list.append('N/A')
    driver.implicitly_wait(2)
    
    # Speed
    try:
        speed = driver.find_element_by_xpath('//*[@class="pokemon__stat__key"][contains(text(), "Speed")]/following-sibling::div[2]')
        speed_list.append(speed.text)
    except:
        speed_list.append('N/A')
    driver.implicitly_wait(2)
    
    # Total Base Stats
    try:
        total_stats = driver.find_element_by_xpath('//*[@class="pokemon__stat pokemon__stat--total"]/b')
        total_stats_list.append(total_stats.text)
    except:
        total_stats_list.append('N/A')
    driver.implicitly_wait(2)
    
    
```

```python
driver.close()
```

```python
pokemon_links[0:10]
```

```python
# Change to title case for pokemon types:

poke_type_one = [x.title() for x in poke_type_one]
poke_type_two = [x.title() for x in poke_type_two]
```

```python
# Change strings into integers:

number_list = [int(x) for x in number_list]
hp_list = [int(x) for x in hp_list]
attack_list = [int(x) for x in attack_list]
defense_list = [int(x) for x in defense_list]

sp_atk_list = [int(x) for x in sp_atk_list]
sp_def_list = [int(x) for x in sp_def_list]
speed_list = [int(x) for x in speed_list]
total_stats_list = [int(x) for x in total_stats_list]
```

```python
## Create dataframe:

pokemon_arceus_legends_df = pd.DataFrame({
    'Number': number_list,
    'Pokemon': pokemon_list,
    'Type One': poke_type_one,
    'Type Two': poke_type_two,
    'HP': hp_list,
    'Attack': attack_list,
    'Defense': defense_list,
    'Sp. Atk': sp_atk_list,
    'Sp. Def': sp_def_list,
    'Speed': speed_list,
    'Total Stats': total_stats_list
})
```

```python
pokemon_arceus_legends_df.to_csv("Pokemon_Legends_Arceus_Stats.csv", index=False)
```
